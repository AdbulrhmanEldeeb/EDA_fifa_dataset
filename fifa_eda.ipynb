* Q. How can we examine the relationships between numerical features in the dataset?
    * A. We can examine the relationships between numerical features using a correlation matrix heatmap. This visualization shows the pairwise correlations between features, helping to identify how strongly they are related.

```python
corr_matrix=df.drop(columns=['Name','ID','Club','Position','Joined','Nationality','Preferred Foot','Contract Valid Until']).corr()
plt.figure(figsize=(14,9))
sns.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix of Numerical Features')

plt.show() 
```

* Q. What does the correlation matrix tell us about the players' attributes?
   * A. The correlation matrix shows how strongly different attributes of players are related to each other. Here’s a simple breakdown:
    
    * Age vs. Overall: There is a moderate positive relationship (0.45). Older players tend to have a better overall rating, but it’s not a strong link.
    
    * Overall vs. Potential: A strong positive relationship (0.66). Players with higher overall ratings generally have higher potential ratings.
    
    * Value vs. Overall: A strong positive relationship (0.63). Players with higher overall ratings usually have higher market values.
    
    * Wage vs. Value: A very strong positive relationship (0.86). Higher wages are strongly associated with higher player values.
    
    * International Reputation vs. Value: A strong positive relationship (0.66). Players with a higher international reputation often have higher market values.
    
    * Skill Moves vs. Height/Weight: Skill moves have a negative correlation with height (-0.42) and weight (-0.35). Players with more skill moves tend to be shorter and lighter.
    
    * Height vs. Weight: There is a strong positive relationship (0.75). Taller players are generally heavier.
    
    * Release Clause vs. Value: A very strong positive relationship (0.97). The release clause is almost perfectly correlated with the player’s value. Higher values usually mean higher release clauses.
 
    * Age vs. Potential: A negative correlation (-0.25) exists between age and potential. This means that as players get older, their potential rating tends to decrease. This is expected, as younger players typically have more potential for growth.

    * Age vs. Wage: A positive but weak correlation (0.14) between age and wage suggests that older players have slightly higher wages, though the relationship is not strong.

    * Overall vs. Wage: There is a strong positive correlation (0.57) between overall rating and wage. Players with higher overall ratings generally earn higher wages.

    * Potential vs. Wage: A moderate positive correlation (0.49) indicates that players with higher potential ratings tend to earn higher wages, though not as strongly as with overall rating.

    * International Reputation vs. Age: A moderate positive correlation (0.25) suggests that players with higher international reputations are slightly older.

    * Skill Moves vs. Overall/Value: Skill moves have a moderate positive correlation with overall rating (0.41) and a weaker positive correlation with value (0.32). Players with better skill moves tend to have higher overall ratings and values, but the relationships are not very strong.

    * Height vs. Age: A weak positive correlation (0.08) between height and age suggests that height does not change significantly with age in this dataset.
    
    * Height vs. Release Clause: A very weak positive correlation (0.0028) between height and release clause indicates almost no relationship between these attributes.

    * Weight vs. Age: There is a weak positive correlation (0.23) between weight and age, suggesting that weight may slightly increase with age, but the relationship is not strong.
    
* In Summary:
    
        Higher overall ratings are linked with higher potential, value, and wages.
        Skill moves and physical attributes (height and weight) show some interesting relationships.
        Release clause and value are almost directly related.
        This matrix helps us understand the key relationships between different attributes of players, making it easier to see how these features interact with each other.

```python
st_players_df=df[df['Position']=='ST']
st_players_df.head(3)
```

```python
def plot_top_10_bar(column,color): 
    # Sort and plot the top 10 players by 'Overall' rating
    ax = st_players_df.sort_values(column, ascending=False)[['Name', column]][:10].set_index('Name').plot(kind='barh', figsize=(10, 7), color=color)
    plt.xlabel(column)
    plt.title(f'Top 10 Strikers by {column}')
    
    # Annotate the values on the bars
    for index, value in enumerate(st_players_df.sort_values(column, ascending=False)[column][:10]):
        ax.text(value, index, str(value), va='center')
    
    plt.show()
        
```
